```python
import pandas as pd

# 1. Configurações de Conexão
sheet_id = "1ZPkLoWjcbtqkQxLL7XHVupH1orXDtKqP4f7l-BOOiRs"
sheet_name = "produtos-eflorista-2025-10-23"
url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"

def exibir_produtos(df, coluna, termo):
    """Função auxiliar para filtrar e exibir produtos"""
    filtro = df[df[coluna].str.contains(termo, case=False, na=False)]
    if not filtro.empty:
        print(f"\n✅ Encontrei esses itens para '{termo}':")
        for _, linha in filtro.iterrows():
            print(f"👉 {linha['Nome do Produto']} - {linha['Venda']}")
    else:
        print(f"\n❌ Poxa, no momento não temos produtos para '{termo}'.")

try:
    df = pd.read_csv(url)

    print("--- 🌸 Floricultura Brasília: Sistema Iniciado ---")
    print("\nBot: Olá! Seja muito bem-vindo(a) à nossa Floricultura! 🌿")
    print("Bot: Sou seu assistente virtual e estou aqui para te ajudar a encontrar o presente perfeito.")

    while True:
        print("\n" + "="*50)
        print("MENU PRINCIPAL")
        print("Como você prefere ver nossas indicações?")
        print("1 - Por Categoria (Tipo de produto)")
        print("2 - Por Ocasião (Para quem ou qual evento)")
        print("0 - Sair do atendimento")

        opcao_principal = input("\nEscolha uma opção: ")

        if opcao_principal == "1":
            while True:
                print("\n--- 📂 CATEGORIAS ---")
                print("1 - Buquês")
                print("2 - Arranjos")
                print("3 - Cestas e Kits")
                print("4 - Plantas")
                print("5 - Coroas")
                print("9 - << Voltar ao menu anterior")

                cat = input("\nQual categoria te interessa? ")

                categorias = {"1": "Buquês", "2": "Arranjos", "3": "Cestas e Kits", "4": "Plantas", "5": "Coroas"}

                if cat in categorias:
                    exibir_produtos(df, 'Categoria', categorias[cat])
                elif cat == "9":
                    break
                else:
                    print("Opção inválida!")

        elif opcao_principal == "2":
            while True:
                print("\n--- 🎈 OCASIÕES ---")
                print("1 - Crianças")
                print("2 - Homenagem e Luto")
                print("3 - Aniversário")
                print("4 - Demonstração de Carinho")
                print("5 - Pedido de Namoro / Romântico")
                print("6 - Agradecimento")
                print("7 - Presentear Mãe")
                print("8 - Presentear Pai")
                print("9 - Formatura")
                print("0 - << Voltar ao menu anterior")

                oca = input("\nPara qual ocasião você busca? ")

                ocasoes = {
                    "1": "Crianças", "2": "Homenagem", "3": "Aniversário",
                    "4": "Carinho", "5": "Romântico", "6": "Agradecimento",
                    "7": "Mãe", "8": "Pai", "9": "Formatura"
                }

                if oca in ocasoes:
                    exibir_produtos(df, 'Ocasiões', ocasoes[oca])
                elif oca == "0":
                    break
                else:
                    print("Opção inválida!")

        elif opcao_principal == "0":
            print("\nBot: Foi um prazer te ajudar! Se precisar de mais flores, é só chamar. 👋")
            break
        else:
            print("\nBot: Ops, não entendi. Escolha 1, 2 ou 0.")

except Exception as e:
    print(f"Erro ao carregar sistema: {e}")
```
